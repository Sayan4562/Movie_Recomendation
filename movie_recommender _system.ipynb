```python
v=data['vote_count']
r=data['vote_average']
c=data['vote_average'].mean()
m=data['vote_count'].quantile(.70)
```

```python
data['weighted_average']=((r*v)+(c*m))/(v+m)
```

```python
### Normalize weighted average and Popularity

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scal_vale=scaler.fit_transform(data[["weighted_average","popularity"]])
normalize_dataFrame=pd.DataFrame(scal_vale,columns=["weighted_average","popularity"])
data["score"]=normalize_dataFrame["weighted_average"]*0.5 +normalize_dataFrame["popularity"]*0.5
data["score"]
```

```python
data=data[["movie_id","title","overview","keywords","cast","crew","genres",'score']]
```

```python
data.head(2)
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data["genres"][0]
```

```python
import ast
def convert(obj):
    l=[]
    obj=ast.literal_eval(obj)
    for i in obj:
        l.append(i["name"])
    return l
```

```python
data["genres"]=data["genres"].apply(convert)
```

```python
data["keywords"]=data["keywords"].apply(convert)
data["keywords"][0]
```

```python
def fetch_acter(obj):
    l=[]
    r=0
    obj=ast.literal_eval(obj)
    for i in obj:
        if r!=3:
            r+=1
            l.append(i["name"])
    return l
```

```python
data["cast"]=data["cast"].apply(fetch_acter)
```

```python
def fetch_director(obj):
    l=[]
    obj=ast.literal_eval(obj)
    for i in obj:
        if i['job']=='Director':
            l.append(i["name"])
    return l
```

```python
data["crew"]=data["crew"].apply(fetch_director)
data["crew"]
```

```python
data["overview"]=data["overview"].str.split()
```

```python
def ls_st(l):
    r=[]
    for i in l:
        r.append( i.replace(" ",""))
    return r
```

```python
data["N_cast"]=data["cast"].apply(ls_st)
data["N_keywords"]=data["keywords"].apply(ls_st)
data["N_crew"]=data["crew"].apply(ls_st)
data["N_genres"]=data["genres"].apply(ls_st)
```

```python
data['tag']=data["overview"]+data["N_genres"]+data["N_keywords"]+data["N_cast"]+data["N_crew"]
```

```python
nw_data=data[["movie_id","title","tag","score","overview","cast","crew"]]
nw_data["overview"]=nw_data["overview"].apply(lambda x : " ".join(x))
```

```python
nw_data["tag"]=nw_data["tag"].apply(lambda x : " ".join(x))
```

```python
nw_data["tag"]
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
    
```

```python
nw_data["tag"]=nw_data["tag"].apply(stem)
```

```python
nw_data["score"]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vactors=cv.fit_transform(nw_data["tag"]).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity=cosine_similarity(vactors)
```

```python
similarity[0]
```

```python
def marg(list1,list2):
    marge_list=[]
    for i in list1:
        for j in list2:
            marge_list.append([i,j])
    return marge_list
```

```python
def recomand(movie):
    movie_index=nw_data[nw_data['title']==movie].index[0]
    distances=similarity[movie_index]
    moive_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[0:11]  ##  lambda x:x[1] use for sort element in bases of second item of list
    ##  enumerate use for marge index and similarity value of every movies 
    for i in moive_list:
        print(nw_data.iloc[i[0]].title)
      
```

```python
recomand('Spectre')
```
